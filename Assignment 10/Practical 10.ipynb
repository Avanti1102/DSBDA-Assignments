```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('iris.csv')
```

```python
#List down the features and their types:
#To determine the feature types, we can use the dtypes attribute of the DataFrame.
feature_types = df.dtypes
print(feature_types)
```

```python
#Create a histogram for each feature:
#We can create histograms for each feature using the hist() function of the DataFrame.


# Create histograms for each feature
df.hist(figsize=(10, 8))
#plt.tight_layout()
plt.show()
```

```python
#Create a boxplot for each feature:
#To create boxplots for each feature, we can use the boxplot() function of the DataFrame.

# Create boxplots for each feature
#df.boxplot(figsize=(10, 8))
#plt.tight_layout()
#plt.show()
data_to_plot=[df[x] for x in df.columns[1:-1]]
fig, axes = plt.subplots(1,figsize=(12,8))
dp=axes.boxplot(data_to_plot)
```

5.Compare distributions and identify outliers:
By examining the histograms and boxplots, we can compare the distributions of each feature.
Outliers can be identified as data points that fall significantly outside the whiskers of the boxplots or appear as extreme values in the histograms.
