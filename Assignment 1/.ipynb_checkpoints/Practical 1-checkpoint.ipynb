Turn categorical variables into quantitative variables in Python.

To convert categorical variables into quantitative variables, we can use one-hot encoding or label encoding. Let's use label encoding as an example to convert the "gender" variable into a numerical format.

```python
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# Convert 'gender' variable using label encoding
data['gender'] = label_encoder.fit_transform(data['gender'])
```

This will encode the "Sex" variable as 0 for female and 1 for male.
